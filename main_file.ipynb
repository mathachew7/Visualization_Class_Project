#### Coffee consumption trends in the US over years

```python
# Define the countries to plot
countries = ['United States of America']
plt.figure(figsize=(12, 6))

for country in countries:
    if country in consumption_df['Country'].values:
        plt.plot(consumption_df.columns[1:], consumption_df[consumption_df['Country'] == country].values[0][1:], label=country)

plt.xlabel('Year')
plt.ylabel('Consumption (in thousands of 60kg bags')
plt.title('Coffee Consumption Trends in The United States (1990-2019)')
plt.xticks(rotation= 90)
plt.legend()
plt.show()
```

#### Top 5 Countries with the Highest Coffee Consumption in the Latest Year (2019):

```python
# Sort data by 2019 consumption and get the top 5 countries
top_countries_2019 = consumption_df.sort_values(by='2019', ascending=False)
top_countries_2019.head()
```

###  Trends in Coffee Consumption Over the Years in al countries

```python
# Calculate the mean consumption for each year
mean_consumption = consumption_df.mean(numeric_only=True)

# Plot the average consumption trend over the years
plt.plot(mean_consumption.index, mean_consumption.values)
plt.xlabel('Year')
plt.ylabel('Average Consumption ( in thousands of 60kg bag')
plt.title('Average Coffee Consumption Trend Over the Years')
plt.xticks(rotation= 90)
plt.show()
```

### Calculate Yearly Consumption Growth:

```python
# Calculate the yearly consumption growth rate
yearly_growth = mean_consumption.pct_change() * 100  # Calculate as a percentage

# Plot the yearly consumption growth
plt.bar(yearly_growth.index, yearly_growth.values)
plt.xlabel('Year')
plt.ylabel('Yearly Growth Rate (%)')
plt.title('Yearly Growth Rate of Coffee Consumption')
plt.xticks(rotation= 90)
plt.show()
```

###  Identify Countries with the Highest Increase in Consumption:

```python
# Calculate the percentage change in consumption from 1990 to 2019
consumption_df['Change_1990_to_2019'] = ((consumption_df['2019'] - consumption_df['1990']) / consumption_df['1990']) * 100

# Find the top 5 countries with the highest increase in consumption
top_growth_countries = consumption_df.sort_values(by='Change_1990_to_2019', ascending=False)
top_growth_countries.head()
```
